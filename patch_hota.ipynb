```python
import matplotlib.pyplot as plt

# 閾値と対応するHOTAの値をリストとして定義します。
thresholds = [i * 0.1 for i in range(11)]
hota_scores = [54.1, 54.6, 54.9, 55.0, 55.1, 55.3, 55.2, 54.8, 54.2, 53.5, 52.5]

plt.figure(figsize=(10, 6))

# 背景色をグレーに設定
plt.gca().set_facecolor('lightgrey')

# 線の色をオレンジに設定
plt.plot(thresholds, hota_scores, marker='o', linestyle='-', color='orange')

# 各点の横にHOTAの値を表示します。
for i, (thresh, hota) in enumerate(zip(thresholds, hota_scores)):
    plt.text(thresh, hota, f'{hota:.2f}', fontsize=9, ha='left', va='bottom')

# グリッドを追加（横線のみ）
plt.grid(axis='y')

# 軸ラベルとタイトルを追加
plt.xlabel('Threshold')
plt.ylabel('HOTA Score')
#plt.title('HOTA Score vs Threshold')

# グラフを表示
plt.show()
```
